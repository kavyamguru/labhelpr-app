```python
# Date 03.02.2026
# Create a new Next.js application named "labhelpr"
# Uses the latest recommended Next.js setup
npx create-next-app@latest labhelpr

# Navigate into the newly created Next.js project
cd labhelpr

# Starts the Next.js dev server
# App will be available at http://localhost:3000
npm run dev

# Create an apps/api directory for the backend
# -p ensures parent directories are created if missing
mkdir -p apps/api

# Navigate into the backend API folder
cd apps/api

# Create an isolated Python virtual environment named ".venv"
python -m venv .venv

# Activate the virtual environment on macOS/Linux
source .venv/bin/activate

# Install FastAPI (web framework) and Uvicorn (ASGI server)
pip install fastapi uvicorn

# Freeze installed packages into requirements.txt
# Useful for deployments and reproducible environments
pip freeze > requirements.txt

# Import FastAPI framework
from fastapi import FastAPI

# Initialize FastAPI app with a title
app = FastAPI(title="LabHelpr API")

# Health check endpoint
# Accessible at GET /
@app.get("/")
def health_check():
    """
    Simple health check endpoint to verify
    that the API is running correctly.
    """
    return {
        "status": "ok",
        "message": "LabHelpr API running"
    }

# Start the FastAPI development server
# --reload enables auto-restart on code changes
# API will be available at http://127.0.0.1:8000
uvicorn main:app --reload
```
